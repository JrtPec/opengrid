```python
import os, sys, pytz
import time
import inspect
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

script_dir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
# add the path to opengrid to sys.path
sys.path.append(os.path.join(script_dir, os.pardir, os.pardir))
from opengrid.library import config, houseprint, fluksoapi, plotting
from opengrid.library import wundergroundapi as wg
c=config.Config()

key = c.get('Wunderground', 'apikey')

# find tmpo
sys.path.append(c.get('tmpo', 'folder'))
import tmpo

%matplotlib inline
plt.rcParams['figure.figsize'] = 16,8

# path to data
path_to_data = c.get('data', 'folder')
if not os.path.exists(path_to_data):
    raise IOError("Provide your path to the data in your config.ini file.  This is a folder containing a 'zip' and 'csv' subfolder.")
else:
    path_to_csv = os.path.join(path_to_data, 'csv')
    path_to_fig = os.path.join(path_to_data, 'figures')
    if not os.path.isdir(path_to_fig): os.makedirs(path_to_fig)
```

```python
hp = houseprint.load_houseprint_from_file('hp_anonymous.pkl')

#end = pd.Tim
start = pd.Timestamp('20140307')
end = pd.Timestamp('20150304')

end = end.to_datetime()
start = start.to_datetime()
```

```python
print start
print end
```

```python
gas_sensors = hp.get_sensors_by_type('gas')
while True:
    errors = 0
    for sensor in gas_sensors:
        try:
            fluksoapi.find_csv(path_to_csv,sensor)
        except:
            gas_sensors.remove(sensor)
            errors += 1
    if errors == 0:
        break

dfcum = fluksoapi.load(path_to_csv, gas_sensors, start, end)
dfcum.index.tz# = None
```

```python
df_temp = wg.get_stored_historic_dayaverage(key=key,city='Leuven',start=start,end=end)
df_heatingdegreedays = wg.get_stored_historic_dayaverage(key=key,city='Leuven',start=start,end=end,prop='heatingdegreedays')
df_coolingdegreedays = wg.get_stored_historic_dayaverage(key=key,city='Leuven',start=start,end=end,prop='coolingdegreedays')
df_temp.index.tz = 'Europe/Brussels'
df_heatingdegreedays.index.tz = 'Europe/Brussels'
df_coolingdegreedays.index.tz = 'Europe/Brussels'
```

```python
df_tot = dfcum.join(df_temp)
df_tot = df_tot.join(df_heatingdegreedays)
df_tot = df_tot.join(df_coolingdegreedays)
```

```python
dfinterp = df_tot.resample('d')
```

```python
weathercolumns = set(["T_out","heatingdegreedays","coolingdegreedays"])
```

```python
for sensor in dfinterp:
    if sensor in weathercolumns: continue
    fig,ax1 = plt.subplots()
    ax1.bar(dfinterp.index,dfinterp[sensor],alpha=0.5,align="center")
    plt.ylabel('P [Wh]')
    ax2 = ax1.twinx()
    temp = dfinterp['T_out'].ix[dfinterp[sensor].first_valid_index():dfinterp[sensor][::-1].first_valid_index()]
    heatingdegreedays = dfinterp['heatingdegreedays'].ix[dfinterp[sensor].first_valid_index():dfinterp[sensor][::-1].first_valid_index()]
    coolingdegreedays = dfinterp['coolingdegreedays'].ix[dfinterp[sensor].first_valid_index():dfinterp[sensor][::-1].first_valid_index()]

    ax2.plot_date(temp.index,temp,'-',color='r',label='Outside Temperature')
    ax2.plot_date(heatingdegreedays.index,heatingdegreedays,'-',color='g',label='Heating Degree Days')
    ax2.plot_date(coolingdegreedays.index,coolingdegreedays,'-',color='y',label='Cooling Degree Days')


    plt.ylabel('T [C]')
    plt.legend()
    plt.title(hp.get_flukso_from_sensor(sensor))
```

```python
for sensor in dfinterp:
    if sensor in weathercolumns :continue
    y = dfinterp[sensor]
    x = dfinterp['T_out']
    
    X = []
    Y = []
    for xval,yval in zip(x,y):
        if np.isnan(xval) or np.isnan(yval): continue
        X.append(xval)
        Y.append(yval)

    Xcold = []
    Ycold = []
    Xnorm = []
    Ynorm = []
    Xwarm = []
    Ywarm = []
    for xval,yval in sorted(zip(X,Y)):
        if xval <= 20:
            Xcold.append(xval)
            Ycold.append(yval)
            '''
        if 14 <= xval <= 21:
            Xnorm.append(xval)
            Ynorm.append(yval)
            '''
        if 19 <= xval:
            Xwarm.append(xval)
            Ywarm.append(yval)
    
    if not len(Xcold) == 0:
        zcold = np.polyfit(Xcold, Ycold, 1)
        pcold = np.poly1d(zcold)
        '''
    if not len(Xnorm) == 0:
        znorm = np.polyfit(Xnorm, Ynorm, 1)
        pnorm = np.poly1d(znorm)
        '''
    if not len(Xwarm) == 0:
        zwarm = np.polyfit(Xwarm, Ywarm, 1)
        pwarm = np.poly1d(zwarm)

    plt.figure()
    colors = []
    for date in y.index:
        delta1 = abs((date - date.replace(month=1,day=10)).days)
        delta2 = abs((date - date.replace(year=date.year+1,month=1,day=10)).days)
        delta = min(delta1,delta2)
        colors.append(delta)
    area = 100# 0 to 15 point radiuses
    
    plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap=plt.get_cmap('summer'))
    
    if not len(Xcold) == 0:
        plt.plot(Xcold,pcold(Xcold),"b-")
        '''
    if not len(Xnorm) == 0:
        plt.plot(Xnorm,pnorm(Xnorm),"y-")
        '''
    if not len(Xwarm) == 0:
        plt.plot(Xwarm,pwarm(Xwarm),"r-")
    
    plt.title(hp.get_flukso_from_sensor(sensor))
    plt.ylabel('P [Wh]')
    plt.xlabel('T [C]')
    
    plt.savefig(os.path.join(path_to_fig, 'gas_outsidetemp'+hp.get_flukso_from_sensor(sensor)+'_tmpo_'+sensor), dpi=100)
```
