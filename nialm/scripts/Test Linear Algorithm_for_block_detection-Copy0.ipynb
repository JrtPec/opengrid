```python
blocks = []
for i in range(len(problem.events)):
    event = match(events=problem.events,i=i)
    if event:
        block = Block(event,problem.events[i])
        if block.is_valid(norm_signal=problem.df_norm):
            blocks.append(block)
            #block.remove_block_from_signal(norm_data=problem.df_norm)
drawables=[]
for block in blocks:
    drawable = block.get_drawable()
    drawables.append(drawable)
    
plt.figure()
plt.plot_date(problem.df_orig.index,problem.df_orig,'-')
for drawable in drawables:
    plt.plot_date(drawable.index,drawable,'-')
    
```

```python
import random
all_blocks = blocks
class Ind(object):
    
    def __init__(self,norm_data,blocks=None,init=True):
        self.data = norm_data.copy()
        self.blocks = []
        if init == True:
            self.init_blocks()
        else:
            self.re_init_blocks(blocks)
                
    def init_blocks(self):
        for block in random.sample(all_blocks,len(all_blocks)):
            if block.is_valid(norm_signal=self.data):
                self.blocks.append(block)
                block.remove_block_from_signal(norm_data=self.data)
                
    def re_init_blocks(self,blocks):
        for block in blocks:
            if block is not None and block.is_valid(norm_signal=self.data):
                self.blocks.append(block)
                block.remove_block_from_signal(norm_data=self.data)
        self.init_blocks()
                
    def get_score(self):
        score = 0.0
        for block in self.blocks:
            score += block.score
        return score / problem.score_ceiling
```

```python
from itertools import izip_longest
class Pop(object):
    def __init__(self,size):
        self.inds = []
        for i in range(0,size):
            ind = Ind(norm_data=problem.df_norm)
            self.inds.append(ind)
            
    def sort(self):
        self.inds = sorted(self.inds, key=lambda Ind: Ind.get_score())[::-1]
        
    def evolve(self):
        for i,ind in enumerate(self.inds):
            if i < 1: continue
            parentA,parentB = random.sample(self.inds[:10],2)
            self.inds[i] = self.breed(parentA,parentB)
            
    def breed(self,parentA,parentB):
        blocks = []
        for blockA,blockB in izip_longest(parentA.blocks,parentB.blocks,fillvalue=None):
            blocks.append(random.choice([blockA,blockB]))
        return Ind(blocks=blocks,norm_data=problem.df_norm,init=False)
```

```python
pop = Pop(size=20)

for i in range(0,10):
    pop.sort()
    if pop.inds[0].get_score() >= 1:
        break
    pop.evolve()
    
```

```python
pop.sort()
drawables=[]
for block in pop.inds[0].blocks:
    drawable = block.get_drawable()
    drawables.append(drawable)
    
plt.figure()
plt.plot_date(problem.df_orig.index,problem.df_orig,'-')
#plt.plot_date(problem.df_norm.index,problem.df_norm,'-')
for drawable in drawables:
    plt.plot_date(drawable.index,drawable,'-')
```
