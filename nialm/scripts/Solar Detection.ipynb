```python
import pandas as pd
import time, os, sys, inspect
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.rcParams['figure.figsize'] = 16,8
```

```python
script_dir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
sys.path.append(os.path.join(script_dir, os.pardir, os.pardir))
sys.path.append(os.path.join(script_dir, os.pardir, os.pardir,'..'))
```

```python
start = pd.Timestamp(time.time() - 3600*72, unit='s') #24 hours ago
start = start.replace(hour=6,minute=0, second=0) #start around 20h
end = start.replace(hour=18,minute=0,second=0)
```

```python
from nialm.library.get_data import get_raw_data
from opengrid.library import houseprint
```

```python
df = get_raw_data(start,end)
try:
    hp = houseprint.load_houseprint_from_file('hp_anonymous.pkl')
except:
    raise IOError("Houseprint file not found")
```

```python
newindex = df.resample('s').index

#Add both the old time index and the new time index together
df = df.reindex(df.index + newindex)

#Interpolate data
df = df.interpolate(method='time')

#resample and take diff
df = df.resample('min').diff()*3600/60
```

```python
def get_solar_for_sensor(sensor):
    for flukso,sensors in hp.fluksosensors.items():
        if hp.get_flukso_from_sensor(sensor) in (None,flukso):
            for int_sensor,sensordic in sensors.items():
                if sensordic is not None and 'electricity' in sensordic['Type'] and 'Solar' in sensordic['Function']:
                    return sensordic['Sensor']
def is_main_sensor(sensor):
    for flukso,sensors in hp.fluksosensors.items():
        if hp.get_flukso_from_sensor(sensor) in (None,flukso):
            for int_sensor,sensordic in sensors.items():
                if sensordic is not None and 'electricity' in sensordic['Type'] and 'Main' in sensordic['Function']:
                    return True
    return False
```

```python
main_sensors = []
solar_sensors = []
for sensor in df:
    if is_main_sensor(sensor) and get_solar_for_sensor(sensor) is not None:
        main_sensors.append(sensor)
        solar_sensors.append(get_solar_for_sensor(sensor))
```

```python
op = []
for main,solar in zip(main_sensors,solar_sensors):
    diff = df[solar] - df[main]
    diff = diff[diff>0]
    
    overproduction = round(diff.sum()/60/1000,3)
    op.append(overproduction)
    
    solar_pos = df[solar].reindex(diff.index).resample('min')
    diff = diff.resample('min').fillna(0)
    
    plt.figure()
    ax1 = plt.subplot(121)
    plt.plot_date(df[main].index,df[main],'-',label='main')
    plt.plot_date(df[solar].index,df[solar],'-',label='solar')
    plt.plot_date(solar_pos.index,solar_pos,'-',label='overproduction')
    plt.legend()
    plt.title(df[main].name)
    ax2 = plt.subplot(122)
    plt.plot_date(df[main].index,df[main],'-',label='main')
    plt.plot_date(diff.index,diff,'-',label='nett overproduction',color='r')
    plt.title(str(overproduction)+"kWh")
    plt.legend()
```
