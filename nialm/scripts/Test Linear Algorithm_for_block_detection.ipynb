```python
import pandas as pd
import time, os, sys, inspect
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.rcParams['figure.figsize'] = 16,8
```

```python
script_dir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
sys.path.append(os.path.join(script_dir, os.pardir, os.pardir))
```

```python
start = pd.Timestamp(time.time() - 3600*24, unit='s') #24 hours ago
start = start.replace(hour=20,minute=0, second=0) #start around 20h
end = start.replace(hour=20,minute=30,second=0)
```

```python
from nialm.library.get_data import get_flukso_data
```

```python
df = get_flukso_data(start,end,sensor='214ef2d3533899dbf633caeb32e80483',token='7d964f8a361a1273594161c67a8ac31e')
```

```python
from nialm.library.problems import *
#pick a sensor
#sensor = '214ef2d3533899dbf633caeb32e80483'
#problem = Disaggregation_problem(orig_data=df[sensor],mutation_rate=0.1)
problem = Disaggregation_problem(orig_data=df,mutation_rate=0.05)

plt.figure()
plt.plot_date(problem.df_orig.index,problem.df_orig,'-',label="Instantaneous Power")
plt.plot_date(problem.df_norm.index,problem.df_norm,'-',label="Normalized Instantaneous Power")
plt.plot_date(problem.df_diff_full.index,problem.df_diff_full,'-',label="1st Derivative")
plt.figtext(0.5,0.95,["Detection Ceiling = ",problem.score_ceiling," Wh"], fontsize=14, ha='center')
plt.legend()
```

```python
from nialm.library.block_detection_methods3 import *
from nialm.library.block import Block
```

```python
blocks = []
for i in range(len(problem.events)):
    event = match(events=problem.events,i=i)
    if event:
        block = Block(event,problem.events[i])
        if block.is_valid(norm_signal=problem.df_norm):
            blocks.append(block)
            #block.remove_block_from_signal(norm_data=problem.df_norm)
drawables=[]
for block in blocks:
    drawable = block.get_drawable()
    drawables.append(drawable)
    
plt.figure()
plt.plot_date(problem.df_orig.index,problem.df_orig,'-')
plt.plot_date(problem.df_norm.index,problem.df_norm,'-')
for drawable in drawables:
   plt.plot_date(drawable.index,drawable,'-')
    
```

# Create Scatter Graph of found blocks

```python
x = []
y = []
for part in blocks:
    x.append(part.get_duration())
    y.append(part.get_avg_power())

colors = np.random.rand(len(blocks))

plt.figure()
plt.scatter(x,y, s=200,c=colors, alpha=0.5)
plt.xlabel('Duration [s]')
plt.ylabel('Power [W]')
plt.grid()
plt.show()
```
